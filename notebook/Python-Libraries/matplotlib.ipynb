# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting inn Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup Plot
# figsize is the x and y axis range
fig, ax = plt.subplots(figsize=(10, 10)) # figsize(width, height)

# 3. Plot Data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
        xlabel = "x_axis",
        ylabel = "y_axis")

# 5. Save the data
fig.savefig("images/sample-plot.png")
```

## Bar plot

```python
# make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter" : 12}
fig, ax = plt.subplots()

ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title = "Butter Store",
      ylabel = "Price ($)",
      xlabel = "Butter");
```

```python
# barh is to represent the histogram in horizontal form.
# The values that are passed in barh needs to be in the form of list.
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

## Histogram plot

```python
# Make some data
# Histogram needs only one value
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

# Two options for subplots

### Subplot option 1

```python
# Subplot option 1
# The ax(n) is the number of graphs or subplots needed
# Histogram needs only one value
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10, 5))

# plot to each different axis
ax0.plot(x, x/2);
ax1.scatter(np.random.rand(10), np.random.rand(10));
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.hist(np.random.random(10));
```

### Subplot option 2

```python
# Subplots option 2
# Axis are taken based on the row and column present
fig, ax = plt.subplots(nrows = 2, 
                      ncols = 2, 
                      figsize = (10, 5))

# Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.rand(10), np.random.rand(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.random(10));
```

### Which is better to use (pyplot (or) matplotlib OO method)

* Simple usecases pyplot can be used
* For advance plotting OO is better

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

## pyplot method

```python
# c is for color
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target',
            figsize=(10, 6));
```

## OO method

```python
# OO method with pyplot
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45, 100])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Adding Legend
ax.legend(*scatter.legend_elements(), title="target");
# Adding a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Creating plots for age, chol, thalach

fig, (ax0, ax1) = plt.subplots(nrows = 2, 
                              ncols = 1,
                              figsize=(10, 10),
                              sharex = True)

# Add data to ax0

scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])

# Adding title
ax0.set(title = "Heart Disease and Cholesterol Levels",
        ylabel = "Cholesterol");

# Adding legend to ax0
ax0.legend(*scatter.legend_elements(), title = "target")

# Adding meanline
ax0.axhline(y = over_50["chol"].mean(),
           linestyle = "--");

# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])

# Adding title
ax1.set(title = "Heart Disease and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate");

# Adding legend to ax1
ax1.legend(*scatter.legend_elements(), title = "target")

# Adding meanline
ax1.axhline(y = over_50["thalach"].mean(),
           linestyle = "--");

# Adding title to the whole figure(plot)
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold");
```
